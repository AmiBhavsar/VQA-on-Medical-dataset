```python
import pandas as pd

# Read the text file
with open(r"D:\Med VQA\Training Data\ImageClef-2019-VQA-Med-Training\ImageClef-2019-VQA-Med-Training\All_QA_Pairs_train.txt", 'r') as file:
    lines = file.readlines()

# Split the text using the "|" separator and create a list of lists
data = [line.strip().split('|') for line in lines]

# Create a DataFrame
df = pd.DataFrame(data)

# Optionally, assign column names to the DataFrame
# Example:
# df.columns = ['Column1', 'Column2', 'Column3', ...]

# Display the DataFrame
print(df)
```

```python
df.rename(columns={0:"ID",1:"Question",2:"Answer"},inplace=True)
df.head()
```

```python
df.shape
```

```python
def determine_answer_type(answer):
    if answer.lower() in ['yes', 'no']:
        return 'Closed'
    else:
        return 'Opened'

# Apply the function to create the Answer_type column
df['Answer_type'] = df['Answer'].apply(determine_answer_type)

# Display the DataFrame
df
```
